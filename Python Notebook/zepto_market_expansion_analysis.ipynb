```python
df1.columns=['city','zepto','blinkit','instamart']
```

```python
df1.head()
```

```python
# Remove % symbol
df1["zepto"] = df1["zepto"].str.replace("%", "")
df1["blinkit"] = df1["blinkit"].str.replace("%", "")
df1["instamart"] = df1["instamart"].str.replace("%", "")

# Convert to numeric
df1["zepto"] = pd.to_numeric(df1["zepto"], errors="coerce")
df1["blinkit"] = pd.to_numeric(df1["blinkit"], errors="coerce")
df1["instamart"] = pd.to_numeric(df1["instamart"], errors="coerce")

df1.head()
```

```python
df1=df1.fillna(0)
```

```python
df1.tail()
```

```python
df1["Competitor_Total"] = df1["blinkit"] + df1["instamart"]
df1["zepto_Gap"] = df1["zepto"] - df1["Competitor_Total"]

df1.head()
```

```python
df1.tail()
```

```python
# Remove rows where all platforms are zero
df1 = df1[
    (df1['zepto'] != 0) |
    (df1['blinkit'] != 0) |
    (df1['instamart'] != 0)
]

df1 = df1.reset_index(drop=True)

df1.head()
```

```python
df1.shape
```

```python
# Top 10 High Competition Cities

high_comp = df1.sort_values(
    by='Competitor_Total',
    ascending=False
).head(10)

high_comp
```

```python
weak_markets = df1.sort_values(
    by='zepto_Gap'
).head(10)

weak_markets
```

```python
# Normalize competitor demand
df1['DemandScore'] = df1['Competitor_Total'] / df1['Competitor_Total'].max()

# Normalize weakness (convert negative gap to positive score)
df1['WeaknessScore'] = abs(df1['zepto_Gap']) / abs(df1['zepto_Gap'].min())

# Final Opportunity Score
df1['OpportunityScore'] = (
    0.6 * df1['DemandScore'] +
    0.4 * df1['WeaknessScore']
)

# Rank cities
final_rank = df1.sort_values(
    by='OpportunityScore',
    ascending=False
)

final_rank.head(10)
```

```python
final_rank[['city','OpportunityScore']].head(5)
```

```python
def classify_tier(score):
    if score >= 0.85:
        return "Immediate Expansion"
    elif score >= 0.75:
        return "Short-Term Expansion"
    else:
        return "Monitor"

final_rank['Expansion_Tier'] = final_rank['OpportunityScore'].apply(classify_tier)

final_rank.head(34)
```

```python
metros = [
    "New Delhi", "Mumbai", "Bengaluru", "Chennai",
    "Hyderabad", "Pune", "Kolkata", "Noida", "Gurgaon"
]

true_expansion = final_rank[
    ~final_rank["city"].isin(metros)
]

true_expansion.head(10)
```

```python
expansion_avg = true_expansion[[
    "zepto",
    "Competitor_Total"
]].mean()

expansion_avg
```

```python
from google_play_scraper import reviews,Sort
import pandas as pd
import numpy as np
```

```python
zepto_app = "com.zeptoconsumerapp"
blinkit_app = "com.grofers.customerapp"
swiggy_app = "in.swiggy.android"
```

```python
zepto_reviews,_ = reviews(
    zepto_app,
    lang='en',
    country='in',
    sort=Sort.NEWEST,
    count=1000
)

zepto_df = pd.DataFrame(zepto_reviews)

zepto_df.head()
```

```python
zepto_df = zepto_df[['content', 'score', 'at']]
zepto_df.head()
```

```python
zepto_df.shape
```

```python
blinkit_reviews,_= reviews(
    blinkit_app,
    lang='en',
    country='in',
    sort=Sort.NEWEST,
    count=1000
)

blinkit_df = pd.DataFrame(blinkit_reviews)

blinkit_df.head()
```

```python
blinkit_df = blinkit_df[['content', 'score', 'at']]
blinkit_df.head()
```

```python
instamart_reviews, _ = reviews(
    swiggy_app,
    lang='en',
    country='in',
    count=1000
)

instamart_df = pd.DataFrame(instamart_reviews)

instamart_df.head()
```

```python
instamart_df = instamart_df[['content', 'score', 'at']]
instamart_df.head()
```

```python
blinkit_df.shape
```

```python
instamart_df.shape
```

```python
zepto_df["Company"] = "Zepto"
blinkit_df["Company"] = "Blinkit"
instamart_df["Company"] = "Instamart"
```

```python
all_reviews = pd.concat([zepto_df, blinkit_df, instamart_df], ignore_index=True)

all_reviews.shape
```

```python
def classify_sentiment(score):
    if score >= 4:
        return "Positive"
    elif score == 3:
        return "Neutral"
    else:
        return "Negative"

all_reviews["Sentiment"] = all_reviews["score"].apply(classify_sentiment)
```

```python
all_reviews.head()
```

```python
sentiment_company = (
    all_reviews
    .groupby(["Company", "Sentiment"])
    .size()
    .unstack()
)

sentiment_company_percent = sentiment_company.div(
    sentiment_company.sum(axis=1), axis=0
) * 100

sentiment_company_percent
```

```python
import re
from collections import Counter

# Filter only Zepto negative reviews
zepto_negative = all_reviews[
    (all_reviews["Company"] == "Zepto") &
    (all_reviews["Sentiment"] == "Negative")
]

# Clean text
text = " ".join(zepto_negative["content"].str.lower())

# Remove punctuation
text = re.sub(r'[^a-z\s]', '', text)

# Split words
words = text.split()

# Remove common stopwords
stopwords = set(["the","is","and","to","a","of","in","it","for","on","my","this","that","with","was","very","not","have","but"])
words = [word for word in words if word not in stopwords and len(word) > 3]

# Count frequency
word_counts = Counter(words)

word_counts.most_common(20)
```

```python
def categorize_issue(text):
    text = text.lower()

    if "delivery" in text or "minute" in text or "late" in text:
        return "Delivery Issue"
    elif "refund" in text or "support" in text or "customer" in text:
        return "Support Issue"
    elif "order" in text or "item" in text or "product" in text:
        return "Order/Product Issue"
    else:
        return "Other"

all_reviews["Issue_Category"] = all_reviews["content"].apply(categorize_issue)
```

```python
issue_distribution = (
    all_reviews[all_reviews["Company"] == "Zepto"]
    .groupby("Issue_Category")
    .size()
    .sort_values(ascending=False)
)

issue_distribution
```

```python
zepto_negative = all_reviews[
    (all_reviews["Company"] == "Zepto") &
    (all_reviews["Sentiment"] == "Negative")
]

issue_distribution = (
    zepto_negative
    .groupby("Issue_Category")
    .size()
    .sort_values(ascending=False)
)

issue_distribution
```

```python
issue_percentage = (
    zepto_negative["Issue_Category"]
    .value_counts(normalize=True) * 100
)

issue_percentage
```
